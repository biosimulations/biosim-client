# Create a client, get a run_id, and visualize the data

```python
client = SimdataClient()

run_id = "649b11e033437e21669d5733"
simdata = client.get_simdata(run_id)
for dataset_name in simdata.dataset_names():
    dataset = simdata.get_dataset(dataset_name)
    
    df = dataset.to_pandas()
    nparray = dataset.to_numpy()
    
    print(f"run_id = {run_id}, dataset = {dataset_name}")
    if (dataset.attributes["_type"] == "SedReport"):
        display(df.head())
    if (dataset.attributes["_type"] == "SedPlot2D"):
        df.plot(x=df.columns[0])
        plt.show()
    
```
